```python
def countplot_with_perc(df, colname, hue=False, **kwargs):
    perc = kwargs.get("perc", False)
    title = kwargs.get("title", False)
    shift = kwargs.get("shift", 2)
    if hue:
        ax = sns.countplot(x=colname, hue=hue, data=df)
        col_vals = df[colname].unique()
        totals = []
        for c in col_vals:
            totals.append(float(len(df[df[colname] == c])))
    else:
        col_vals = ['total']
        ax = sns.countplot(x=colname, data=df)
        total = float(len(df))
        totals = [total]
    totals = totals[::-1]
    if perc:
        for i, p in enumerate(ax.patches):
            if len(col_vals) > 1:
                total = totals[i % len(col_vals)]
            height = p.get_height()
            ax.text(
                p.get_x() + p.get_width() / 2.,
                height - shift,
                '{:1.2f} %'.format(100 * height / total),
                ha="center",
                color="black",
                size=10)
    if title:
        sns.plt.title(title)
    return ax


def table_version(table, title, caption, f=None):
    """Wrap obj.to_latex() into latex table with title and caption"""
    global version
    if version == "html":
        return table
    else:
        latex = r"""
\begin{{table}}[h!]
\caption {{{}}} 
\label{{tab:title}} 
\begin{{center}}
{}\end{{center}}
\end{{table}}
""".format(title + caption, table.to_latex(column_format=f) )
        return Latex(latex)

def setup_df(df):
    """ustawienie typów danych"""
    if df.sx.dtype.char != "category":
        df['sx'] = df.sx.astype("category")
        df['yr'] = df['yr'].map(int)
        df['yd'] = df['yd'].map(int)
        df['sl'] = df['sl'].map(int)
        df['rk'] = df['rk'].map({'assistant': 1, 'associate': 2, 'full': 3})
        df['dg'] = df['dg'].map({'masters': 0, 'doctorate': 1})

    return df
```
